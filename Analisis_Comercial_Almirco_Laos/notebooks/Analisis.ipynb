```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
import os
```

```python
base = "VideoAnalisis/documents"
df_ventas = pd.read_csv(os.path.join(base, "Ventas.csv"))
df_inventario = pd.read_csv(os.path.join(base, "Inventario.csv"))
df_productos = pd.read_csv(os.path.join(base, "Productos.csv"))
df_tienda = pd.read_csv(os.path.join(base, "Tiendas.csv"))
```

```python
df_ventas.head()
```

```python
df_inventario.head()
```

```python
df_productos.head()
```

```python
df_tienda.head()
```

```python
df_analisis = df_ventas.merge(df_inventario, on = "ProductoID", how= "left")
```

```python
df_analisis
```

```python
df_analisis = df_analisis.merge(df_tienda, on = "TiendaID", how= "left")
```

```python
df_analisis
```

```python
df_analisis = df_analisis.merge(df_productos, on = ["ProductoID", "PrecioUnitario"], how= "left")
```

```python
df_analisis
```

```python
df_analisis.info()
```

```python
df_analisis["Fecha"] = pd.to_datetime(df_analisis["Fecha"])
```

```python
df_analisis.info()
```

```python
df = df_analisis.copy()
```

```python
df["Mes_numero"] = df["Fecha"].dt.month
df["Mes_nombre"] = df["Fecha"].dt.month_name()
```

```python
df
```

```python
ingresos_por_mes = df.groupby(["Mes_nombre", "Canal"])["Ingreso"].sum().reset_index()
```

```python
ingresos_por_mes
```

```python
ingresos_por_mes = ingresos_por_mes.rename(columns={"Mes_nombre" : "Mes"})
```

```python
ingresos_por_mes
```

```python
df
```

```python
top_10_productos_ingresos = df.groupby("NombreProducto")["Ingreso"].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
top_10_productos_ingresos
```

```python
top_10_productos_unidades = df.groupby("NombreProducto")["Cantidad"].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
top_10_productos_unidades
```

```python
df["Margen"] = df["PrecioUnitario"] - df["CostoUnitario"]
```

```python
Margen = df.groupby("Categoría")["Margen"].mean().reset_index().round(2)
```

```python
Margen
```

```python
df.loc[df["StockActual"] < 50, ["NombreProducto", "NombreTienda", "StockActual"]].drop_duplicates().sort_values("StockActual", ascending = False)
```

```python
df.groupby(["NombreTienda", "Ciudad", "Mes_nombre"])["Ingreso"].sum().reset_index().sort_values(by = "Ingreso", ascending = True).head(5)
```

```python
df.groupby(["Mes_nombre", "Canal"])["Ingreso"].sum().reset_index().sort_values(by = "Ingreso", ascending = True)
```

```python
df.to_csv("Analisis_PT.csv", index = False, encoding = "utf-8-sig")
```

```python
stock_bajo = (
    df.loc[df["StockActual"] < 50, ["NombreProducto", "StockActual"]]
    .drop_duplicates()
    .sort_values("StockActual", ascending=False)
)

stock_bajo.to_csv("Stock_Bajo.csv", index=False, encoding="utf-8-sig")
```
